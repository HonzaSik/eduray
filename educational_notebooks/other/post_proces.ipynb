```python
from eduray import *
```

# Simple scene setup

```python
camera = PinholeCamera(
    origin=Vertex(0, 0, 4),
    direction=Vector(0, 0, -1),
    fov_deg=40.0,
)

light = PointLight(
    position=Vertex(5, 5, 5),
    intensity=1.0,
    color=Color.custom_rgb(255, 255, 255),
)

my_sphere = Object(
    geometry=Sphere(),
    material=PhongMaterial(base_color=Color.custom_rgb(255, 0, 0)),
)

scene=Scene(
    camera=camera,
    lights=[light],
    objects=[my_sphere],
)
```

# Config objects with upscaling

```python
post_process_config = PostProcessConfig(
    enabled = True,
    scale_factor = 4,
)

preview_config = PreviewConfig(
    progress_display=ProgressDisplay.TQDM_IMAGE_PREVIEW
)

render_config = RenderConfig(
    resolution=Resolution.R144p,
    samples_per_pixel=1,
    max_depth=1
)
```

# Render

```python
rt = LinearRenderLoop(
    post_process_config = post_process_config,
    preview_config = preview_config,
    render_config = render_config,
    scene = scene
)

img = rt.render("upscale_4times_test.png")
ipynb_display_images(img)
```
